```python
import requests
from bs4 import BeautifulSoup

#Speakers
speakers_url = 'https://www.knowledgegraph.tech/speakers/'
# Presentations: 
program_url = 'https://www.knowledgegraph.tech/kgc2021/program/#1616729201427-a18317f3-52cd'

# Package the request, send the request and catch the response: r
r=requests.get(speakers_url)

# Extracts the response as html: html_doc
html_doc=r.text

# Create a BeautifulSoup object from the HTML: soup
soup=BeautifulSoup(html_doc)


# Get the title of webpage: title
title=soup.title

# Print the title
print(title)

# Get text: text
text=soup.get_text()

# Print text
# print(text)

#Get hyperlinks

# Find all 'a' tags (which define hyperlinks): a_tags
a_tags=soup.find_all('a')

# Print All the URLs on the Website
for link in a_tags:
    print(link.get('href'))
    
#Some are under keynote and other under speakers    
# That would pull up the specific speaker's page from which we pull their affiliation and organization
# e.g., https://www.knowledgegraph.tech/speakers/denis-streitmatter/
#This is just an example
denis_url = 'https://www.knowledgegraph.tech/speakers/denis-streitmatter/' 
r=requests.get(denis_url)
html_doc=r.text
soup=BeautifulSoup(html_doc)

# Prettify the BeautifulSoup object: pretty_soup
#This will give an idea about the sturcture of the html file
#  <div class="title">
#                Denis Streitmatter
#               </div>
#               <div class="job-position">
#                InfAI at Leipzig University, Germany
#               </div>
#               <div class="jx-evont-social">
#                <ul>
#                </ul>
#               </div>
#               <div class="description">
#              <a href="https://www.knowledgegraph.tech/kgc2021/workshops-and-tutorials/#tut-dbpedia">
pretty_soup=soup.prettify()
# print(pretty_soup)

title=soup.title
print(title)
text=soup.get_text()
# print(text)

# You can use findall to look for div tags and look for the classes and selectively choose title, jobposition, and description
div_tags=soup.find_all('div')
for div in div_tags:
    listt=div.get('class')
    if(listt==['title'] or listt==['job-position'] or listt==['description']):
        print(div.get_text())
    
```
